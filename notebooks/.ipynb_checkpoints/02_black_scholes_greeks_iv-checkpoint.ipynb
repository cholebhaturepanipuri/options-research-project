# Notebook 02 — Black-Scholes Pricing and Greeks

Notebook 01 showed payoff **at expiry**. This notebook answers:
> *What is the option worth right now, before expiry?*

Topics:
1. The Black-Scholes formula and its inputs
2. d1 and d2 — what they actually mean
3. Pricing calls and puts
4. Put-call parity
5. Implied volatility
6. The five Greeks
7. Greeks visualised vs spot and time

````python
import sys
sys.path.insert(0, "../src")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

from option_pricing import bs_price, implied_volatility
from greeks import delta, gamma, theta, vega, rho, all_greeks

S     = 19_500
K     = 19_500
r     = 0.065
sigma = 0.15
T     = 30/365

print(f"S={S}  K={K}  r={r:.1%}  sigma={sigma:.0%}  T={T:.4f}yr (30 days)")
````

---
## 1. The Black-Scholes Formula

```
d1 = [ln(S/K) + (r + sigma^2/2) * T] / (sigma * sqrt(T))
d2 = d1 - sigma * sqrt(T)

Call = S * N(d1) - K * exp(-rT) * N(d2)
Put  = K * exp(-rT) * N(-d2) - S * N(-d1)
```

**N(x)** = standard normal CDF — probability that a standard normal variable is <= x.

- **N(d2)** = risk-neutral probability that the call finishes ITM
- **N(d1)** = delta of the call

````python
d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
d2 = d1 - sigma*np.sqrt(T)

print(f"d1 = {d1:.6f}")
print(f"d2 = {d2:.6f}")
print()
print(f"N(d1)  = {norm.cdf(d1):.6f}  <- call delta")
print(f"N(d2)  = {norm.cdf(d2):.6f}  <- P(call finishes ITM)")
print(f"N(-d1) = {norm.cdf(-d1):.6f}  <- |put delta|")
print(f"N(-d2) = {norm.cdf(-d2):.6f}  <- P(put finishes ITM)")
````

---
## 2. Pricing calls and puts

````python
call = bs_price(S, K, r, sigma, T, "call")
put  = bs_price(S, K, r, sigma, T, "put")

print(f"Call price: Rs {call:.2f}")
print(f"Put  price: Rs {put:.2f}")

spots = np.linspace(18_000, 21_000, 300)
call_prices = [bs_price(s, K, r, sigma, T, "call") for s in spots]
put_prices  = [bs_price(s, K, r, sigma, T, "put")  for s in spots]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

for ax, prices, color, label in [
    (ax1, call_prices, "steelblue", "Call"),
    (ax2, put_prices,  "tomato",    "Put"),
]:
    intrinsic = [max(s-K,0) if label=="Call" else max(K-s,0) for s in spots]
    ax.plot(spots, prices,   color=color, lw=2.5, label=f"BS {label} Price")
    ax.plot(spots, intrinsic, color=color, lw=1.5, ls="--", alpha=0.6, label="Intrinsic")
    ax.axvline(K, color="gray", ls="--", alpha=0.7)
    ax.set_title(f"{label} Price vs Spot", fontweight="bold")
    ax.set_xlabel("Spot"); ax.set_ylabel("Price (Rs)")
    ax.legend(); ax.grid(alpha=0.3)

plt.suptitle("BS Price vs Intrinsic — gap is time value", fontsize=12, fontweight="bold")
plt.tight_layout(); plt.show()
````

---
## 3. Put-Call Parity

A model-free relationship that must always hold:

```
Call - Put = S - K * exp(-rT)
```

If this breaks, risk-free arbitrage exists. It always holds in practice.

````python
lhs = call - put
rhs = S - K * np.exp(-r * T)

print(f"Call - Put     = {lhs:.6f}")
print(f"S - K*exp(-rT) = {rhs:.6f}")
print(f"Difference     = {abs(lhs-rhs):.10f}  (should be ~0)")
````

---
## 4. Implied Volatility

Black-Scholes takes vol as **input**, gives price as **output**.

In markets, we observe prices and ask: *what vol does the market imply?*

We solve numerically (Brent's method) for the sigma that makes BS price = market price.

````python
# Round-trip: price at 18% vol, recover it back
market_price = bs_price(S, K, r, 0.18, T, "call")
print(f"Market call price (at 18% vol): Rs {market_price:.2f}")

iv = implied_volatility(market_price, S, K, r, T, "call")
print(f"Recovered IV: {iv:.4f} = {iv:.2%}")
print(f"Error: {abs(iv - 0.18):.2e}")
````

````python
# The vol smile — IV varies by strike
strikes = np.arange(18_500, 20_501, 100)
true_vols = 0.15 + 0.04 * ((strikes - S) / S)**2   # U-shaped smile

prices = [bs_price(S, k, r, v, T, "call") for k, v in zip(strikes, true_vols)]
ivs    = [implied_volatility(p, S, k, r, T, "call") for p, k in zip(prices, strikes)]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(strikes, [v*100 for v in ivs], color="steelblue", lw=2.5, marker="o", ms=4)
ax.axvline(S, color="gray", ls="--", label=f"ATM ({S:,})")
ax.set_title("Implied Volatility Smile", fontsize=13, fontweight="bold")
ax.set_xlabel("Strike"); ax.set_ylabel("IV (%)")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
print("OTM options have higher IV than ATM — the market prices in fat tails.")
````

---
## 5. The Five Greeks

| Greek | What it measures | Call sign | Put sign |
|-------|-----------------|-----------|----------|
| **Delta** | Price change per Rs1 spot move | 0 to +1 | -1 to 0 |
| **Gamma** | Delta change per Rs1 spot move | + | + |
| **Theta** | Price change per 1 day passing | - | - |
| **Vega**  | Price change per 1% vol move | + | + |
| **Rho**   | Price change per 1% rate move | + | - |

````python
g_call = all_greeks(S, K, r, sigma, T, "call")
g_put  = all_greeks(S, K, r, sigma, T, "put")

labels = {
    "delta": "Rs price change per Rs1 spot move",
    "gamma": "delta change per Rs1 spot move",
    "theta": "price decay per calendar day",
    "vega" : "price change per 1% vol move",
    "rho"  : "price change per 1% rate move",
}

print(f"{'Greek':<8} {'Call':>10} {'Put':>10}  Meaning")
print("-" * 60)
for k in g_call:
    print(f"{k:<8} {g_call[k]:>10.5f} {g_put[k]:>10.5f}  {labels[k]}")
````

---
## 6. All Greeks vs Spot

````python
spots = np.linspace(K*0.75, K*1.25, 300)

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes_flat = axes.flatten()

configs = [
    ("Delta", lambda s, ot: delta(s, K, r, sigma, T, ot), True),
    ("Gamma", lambda s, ot: gamma(s, K, r, sigma, T),     False),
    ("Theta", lambda s, ot: theta(s, K, r, sigma, T, ot), True),
    ("Vega",  lambda s, ot: vega(s, K, r, sigma, T),      False),
    ("Rho",   lambda s, ot: rho(s, K, r, sigma, T, ot),   True),
]

for i, (name, fn, has_put) in enumerate(configs):
    ax = axes_flat[i]
    ax.plot(spots, [fn(s, "call") for s in spots], color="steelblue", lw=2, label="Call")
    if has_put:
        ax.plot(spots, [fn(s, "put") for s in spots], color="tomato", lw=2, label="Put")
    ax.axvline(K, color="gray", ls="--", alpha=0.6)
    ax.axhline(0, color="gray", lw=0.7)
    ax.set_title(name, fontsize=12, fontweight="bold")
    ax.set_xlabel("Spot"); ax.legend(fontsize=9); ax.grid(alpha=0.3)

axes_flat[-1].set_visible(False)
plt.suptitle(f"Greeks vs Spot  (K={K:,}, sigma={sigma:.0%}, T=30d)", fontsize=13, fontweight="bold")
plt.tight_layout(); plt.show()
````

---
## 7. Theta decay — time is your edge as a seller

Short options earn theta every day. The decay accelerates near expiry — the last two weeks are the richest for sellers.

````python
days = np.linspace(0.5, 60, 200)

call_prices_t = [bs_price(S, K, r, sigma, d/365, "call") for d in days]
put_prices_t  = [bs_price(S, K, r, sigma, d/365, "put")  for d in days]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(days, call_prices_t, color="steelblue", lw=2.5, label="ATM Call")
ax.plot(days, put_prices_t,  color="tomato",    lw=2.5, label="ATM Put")
ax.invert_xaxis()
ax.set_xlabel("Days to Expiry (right = expiry)")
ax.set_ylabel("Option Price (Rs)")
ax.set_title("Theta Decay — Price vs Days Remaining", fontsize=13, fontweight="bold")
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()
````

---
## 8. ATM call price reference table

How price changes across vol levels and days to expiry.

````python
vols = [0.10, 0.12, 0.15, 0.18, 0.20, 0.25, 0.30]
dtes = [7, 14, 21, 30, 45, 60]

data = {}
for v in vols:
    row = {f"{d}d": round(bs_price(S, K, r, v, d/365, "call"), 1) for d in dtes}
    data[f"{v:.0%} vol"] = row

df = pd.DataFrame(data).T
print("ATM Call Price (Rs) by vol and DTE")
df
````

---
## Key takeaways

1. **BS formula** uses 5 inputs: S, K, r, sigma, T. Volatility is the only unobservable one.
2. **IV** is what the market implies — it's the market's forecast of future vol embedded in option prices.
3. **The smile**: IV is not flat — OTM options price in fat tails.
4. **Short straddle** is long theta + short vega + short gamma. Time is on your side, vol spikes are against you.
5. **Greeks** tell you your risk exposure in real time.
